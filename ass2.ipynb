```python
import numpy as np

def solve_cramers(A, B):
    A = np.array(A, dtype=float)
    B = np.array(B, dtype=float)
    det_A = np.linalg.det(A)
    
    if abs(det_A) < 1e-9:
        return "No unique solution"
    
    n = len(B)
    solutions = []
    
    for i in range(n):
        A_copy = A.copy()
        A_copy[:, i] = B
        det_Ai = np.linalg.det(A_copy)
        solutions.append(det_Ai / det_A)
    
    return solutions
```

```python
A = [[2, -1, 3],
     [1, 1, 1],
     [3, -1, -2]]
B = [9, 6, 2]
print(solve_cramers(A, B))
```
