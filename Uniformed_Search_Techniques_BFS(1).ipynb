
Implementation of Uniformed Search Techniques: Breadth-First Search (BFS)

Graph Class Definition: The code defines a class named Graph to represent a graph using an adjacency list representation. This class will allow users to create graphs and perform operations like adding edges and performing Breadth-First Search (BFS) traversal.

Constructor (__init__): The constructor method initializes the graph as an empty defaultdict where the keys are nodes and the values are lists of adjacent nodes.

addEdge Method: This method allows adding edges to the graph. It takes three arguments: u, v, and weight, representing the nodes connected by the edge and the weight of the edge. It appends the edge (v, weight) to the adjacency list of node u, and vice versa since it's an undirected graph.

dispg Method: This method is used to display the graph, printing out its adjacency list representation.

BFS Method: This method performs a Breadth-First Search traversal starting from a given source node start and aims to find a goal node goal. It initializes a queue (frontier_q) to keep track of the nodes to visit next. It also initializes a visited array to keep track of visited nodes.

BFS Algorithm:

It starts by enqueuing the start node into the queue and marks it as visited.
While the queue is not empty, it dequeues a node s from the queue.
If the dequeued node s is the goal node, it prints that the goal is found and returns.
Otherwise, it iterates over all adjacent nodes of s and enqueues them if they have not been visited yet, marking them as visited.
If the goal is not found after traversing all reachable nodes, it prints that the goal is not found.

```python
from collections import defaultdict
```

```python
class Graph:
    # Constructor
    def __init__(self):
        # default dictionary to store graph
        self.graph = defaultdict(list)

    # function to add an edge to graph
    def addEdge(self, u, v, weight):
        self.graph[u].append((v, weight))
        self.graph[v].append((u, weight))

    def dispg(self):
        print(self.graph.items())

    # Function to print a BFS of graph
    def BFS(self, start, goal):
        # Mark all the vertices as not visited
        print(len(self.graph))
        visited = [False] * (len(self.graph))
        # Create a queue for BFS
        frontier_q = []
        # Mark the source node as visited and enqueue it
        frontier_q.append(start)
        visited[start] = True
        while frontier_q:
            # Dequeue a vertex from queue and print it
            s = frontier_q.pop(0)
            print(s, end=" ")
            if s == goal:
                print("\n Goal found")
                return
            # Get all adjacent vertices of the dequeued vertex s.
            # If adjacent has not been visited, then mark it visited and enqueue it
            for i, j in self.graph[s]:
                if not visited[i]:
                    frontier_q.append(i)
                    visited[i] = True
        print("\n Goal not found")
```

```python
# Create a graph given in the above diagram
g = Graph()
g.addEdge(0, 1, 9)
g.addEdge(0, 2, 5)
g.addEdge(1, 4, 8)
g.addEdge(2, 3, 2)
g.addEdge(3, 4, 4)
g.addEdge(3, 0, 1)
g.addEdge(4, 0, 4)
g.dispg()
print("Following is Breadth First Traversal (starting from vertex 1)")
g.BFS(1, 4)
```
