```python
from collections import defaultdict
```

```python
class Graph:
    def __init__(self):
        self.graph = defaultdict(list)
    def addEdge(self, u, v, weight):
        self.graph[u].append((v, weight))
        self.graph[v].append((u, weight))

    def dispg(self):
        print(self.graph.items())
    def BFS(self, start, goal):
        print(len(self.graph))
        visited = [False] * (len(self.graph))
        frontier_q = []
        frontier_q.append(start)
        visited[start] = True
        while frontier_q:
            s = frontier_q.pop(0)
            print(s, end=" ")
            if s == goal:
                print("\n Goal found")
                return
            for i, j in self.graph[s]:
                if not visited[i]:
                    frontier_q.append(i)
                    visited[i] = True
        print("\n Goal not found")
```

```python
g = Graph()
g.addEdge(0, 1, 9)
g.addEdge(0, 2, 5)
g.addEdge(1, 4, 8)
g.addEdge(2, 3, 2)
g.addEdge(3, 4, 4)
g.addEdge(3, 0, 1)
g.addEdge(4, 0, 4)
g.dispg()
print("Following is Breadth First Traversal (starting from vertex 1)")
g.BFS(1, 4)
```
