```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, Normalizer
data = np.array([[200], [300], [400], [600], [1000]])

print("Original Data:\n", data)
```

```python
minmax = MinMaxScaler()
minmax_scaled = minmax.fit_transform(data)

print("\nMin-Max Normalized Data:")
print(minmax_scaled)
```

```python
standard = StandardScaler()
standard_scaled = standard.fit_transform(data)

print("\nStandardized Data (Mean=0, Std=1):")
print(standard_scaled)
```

```python
data_row = np.array([[220, 300, 400, 600, 1000]])

normalizer = Normalizer()
l2_normalized = normalizer.fit_transform(data_row)

print("\nL2 Normalized Data (Unit Vector):")
print(l2_normalized)
```
