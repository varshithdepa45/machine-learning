```python
import numpy as np

def solve_cramers(A, B):
    det_A = np.linalg.det(A)

    if abs(det_A) < 1e-10:
        print("Determinant of coefficient matrix is zero. Cannot use Cramer's rule.")
        return None

    n = len(A)
    solutions = []

    for i in range(n):
        A_copy = A.copy()
        A_copy[:, i] = B
        det_Ai = np.linalg.det(A_copy)
        x_i = det_Ai / det_A
        solutions.append(x_i)

    return solutions

A = np.array([[2, -1, 3], [1, 2, 1], [1, 1, 1]])
B = np.array([9, 3, 7])

solutions = solve_cramers(A, B)
print(solutions)
```
