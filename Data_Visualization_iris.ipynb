```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris

```

```python
# Load dataset
iris = load_iris()
X = iris.data
y = iris.target
species = iris.target_names



```

```python
# ---------- 1. BAR CHART ----------
# Average petal length per species
avg_petal_length = [np.mean(X[y==i, 2]) for i in range(3)]

plt.figure()
plt.bar(species, avg_petal_length)
plt.xlabel("Species")
plt.ylabel("Average Petal Length")
plt.title("Bar Chart: Average Petal Length by Species")
plt.show()
```

```python
# ---------- 2. PIE CHART ----------
# Species distribution
counts = [sum(y==i) for i in range(3)]

plt.figure()
plt.pie(counts, labels=species, autopct='%1.1f%%')
plt.title("Pie Chart: Species Distribution")
plt.show()

```

```python

# ---------- 3. SCATTER PLOT ----------
# Sepal Length vs Sepal Width
plt.figure()
plt.scatter(X[:,0], X[:,1], c=y)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Scatter Plot: Sepal Length vs Width")
plt.show()

```

```python

# ---------- 4. HISTOGRAM ----------
# Petal Length distribution
plt.figure()
plt.hist(X[:,2])
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.title("Histogram: Petal Length Distribution")
plt.show()
```
